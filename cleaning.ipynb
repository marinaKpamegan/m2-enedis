### Pre-traitement des variables qualitatives

```python
def qualitative_preprocessing(new_data):
    # On remplace des adresses inconnues par la chaîne "inconnu"
    new_data['Identifiant__BAN'] = new_data['Identifiant__BAN'].fillna('inconnu')
    new_data['Nom__rue_(BAN)'] = new_data['Nom__rue_(BAN)'].fillna('inconnu')
    new_data['Adresse_(BAN)'] = new_data['Adresse_(BAN)'].fillna('inconnu')
    new_data['Adresse_brute'] = new_data['Adresse_brute'].fillna('inconnu')
    new_data['Code_INSEE_(BAN)'] = new_data['Code_INSEE_(BAN)'].apply(lambda x: str(x) if pd.notna(x) else x)


    quali_data = new_data.select_dtypes(include=['object'])
    for column in quali_data.columns:
        # Calculate the proportions of each category
        proportions = quali_data[column].value_counts(normalize=True)
        
        # Define a function to impute based on proportions
        def impute_with_proportions(row):
            if pd.isnull(row[column]):
                return proportions.sample(weights=proportions, replace=True).index[0]
            return row[column]
        
        # Apply the function to the column
        quali_data[column] = quali_data.apply(impute_with_proportions, axis=1)
    
    return quali_data
```
